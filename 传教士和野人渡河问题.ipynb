# 传教士和野人渡河问题
3个传教士与3个野人渡河，河岸有一条船，每次最多只能带2个人，河两岸以及船上的野人数目总是不超过传教士的数目(否则不安全，传教士有可能被野人吃掉)。
问：（1）怎样才能让传教士安全渡河。
（2）如果是n个野人与n个传教士渡河，船上最多能乘坐k人（其中n和k由用户输入），怎样才能让传教士安全渡河。


```python
from collections import deque

class state:
    def __init__(self, left_mis, left_savages, right_mis, right_savages, boat_loc):
        self.left_mis = left_mis
        self.left_savages = left_savages
        self.right_mis = right_mis
        self.right_savages = right_savages
        self.boat_loc = boat_loc

    def __eq__(self, other):
        return (self.left_mis == other.left_mis and
                self.left_savages == other.left_savages and
                self.right_mis == other.right_mis and
                self.right_savages == other.right_savages and
                self.boat_loc == other.boat_loc)

    def __hash__(self):
        return hash((self.left_mis, self.left_savages, self.right_mis, self.right_savages, self.boat_loc))

    def __repr__(self) -> str:
        return f"State({self.left_mis}, {self.left_savages}, {self.right_mis}, {self.right_savages}, '{self.boat_loc}')"
    def __copy__(self):
        return state(self.left_mis, self.left_savages, self.right_mis, self.right_savages, self.boat_loc)

class boat:
    def __init__(self, num_mis, num_savages, boat_loc):
        self.num_mis = num_mis
        self.num_savages = num_savages
        self.boat_loc = boat_loc

    def __repr__(self):
        return f"船上有传教士{self.num_mis}人,野人{self.num_savages}人,从{self.boat_loc}出发。"

def caculate(state, boat):
    if state.boat_loc == boat.boat_loc and boat.boat_loc == "l":
        if (state.left_mis - boat.num_mis >= state.left_savages - boat.num_savages or boat.num_mis == 0) and \
           (state.left_mis - boat.num_mis >= 0 and state.left_savages - boat.num_savages >= 0):
            new_state = state.__copy__()
            new_state.left_mis -= boat.num_mis
            new_state.left_savages -= boat.num_savages
            new_state.right_mis += boat.num_mis
            new_state.right_savages += boat.num_savages
            new_state.boat_loc = "r"
            return new_state
    elif state.boat_loc == boat.boat_loc and boat.boat_loc == "r":
        if (state.right_mis - boat.num_mis >= state.right_savages - boat.num_savages or boat.num_mis == 0) and \
           (state.right_mis - boat.num_mis >= 0 and state.right_savages - boat.num_savages >= 0):
            new_state = state.__copy__()
            new_state.left_mis += boat.num_mis
            new_state.left_savages += boat.num_savages
            new_state.right_mis -= boat.num_mis
            new_state.right_savages -= boat.num_savages
            new_state.boat_loc = "l"
            return new_state
    return None

method = [boat(2, 1, "l"), boat(0, 2, "l"), boat(1, 1, "l"), boat(1, 0, "l"), boat(0, 1, "l"),
          boat(2, 1, "r"), boat(0, 2, "r"), boat(1, 1, "r"), boat(1, 0, "r"), boat(0, 1, "r")]

def dfs(initial_state, goal_state):
    open_list = deque([(initial_state, [])])  # 初始化 open 表
    closed_set = set()                        # 初始化 closed 表
    actions = []                              # 初始化动作集合

    while open_list:
        current_state, path = open_list.pop()
        if current_state == goal_state:
            return closed_set, path, actions
        closed_set.add(current_state)
        for move in method:
            next_state = caculate(current_state, move)
            if next_state and next_state not in closed_set:
                open_list.append((next_state, path + [current_state]))
                actions.append(move)  # 记录动作
    return None

# 示例调用
num_mis = int(input("请输入传教士的个数："))
num_savages = int(input("请输入野人的个数："))
initial_state = state(num_mis, num_savages, 0, 0, "l")
gaol_state = state(0, 0, num_mis, num_savages, "r")
solution = dfs(initial_state, gaol_state)
if solution:
    closed_set, path, actions = solution
    for state in path:
        print( state)
    for action in actions:
        print( action)
    
else:
    print("没有找到解决方案")
```
